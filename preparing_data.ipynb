# Import Library

```python
import pandas as pd
```

# Import Dataset

```python
df = pd.read_csv('Dataset/order.csv', encoding='utf-8-sig', delimiter=',')
df
```

# Change CustomerID type

```python
print(df.dtypes)
```

```python
df['CustomerID']=df["CustomerID"].astype("Int64")
df['CustomerID']=df["CustomerID"].astype(object)
```

```python
print(df.dtypes)
```

# Drop Missing Value

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

# Check Duplicate Data

```python
df.duplicated().sum()
```

# Change Date Format

```python
df["InvoiceDate"] = pd.to_datetime(
    df["InvoiceDate"]
)
```

```python
print(df['InvoiceDate'].dtypes)
```

# Add Revenue

```python
df["Revenue"] = df["Quantity"] * df["UnitPrice"]
```

```python
df
```

# Check Return Invoice

```python
returns = df[
    df["InvoiceNo"]
    .astype(str)
    .str.startswith("C")
]

print(returns.shape)
```

# Add Year & Month Column

```python
df["Year"] = df["InvoiceDate"].dt.year

df["Month"] = df["InvoiceDate"].dt.month

df["MonthName"] = (
    df["InvoiceDate"]
    .dt.strftime("%B")
)

df["Quarter"] = (
    df["InvoiceDate"]
    .dt.quarter
)
```

```python
df
```

# Save to CSV

```python
df.to_csv('Dataset/cleaned_order.csv', index=False, encoding='utf-8-sig')
```
