# Import Library

```python
import pandas as pd
```

# Import Data

```python
df = pd.read_csv('cleaned_order.csv', encoding='utf-8-sig', delimiter=',')
df
```

```python
print(df.dtypes)
```

```python
df['CustomerID']=df["CustomerID"].astype("Int64")
df['CustomerID']=df["CustomerID"].astype(object)
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
```

# Customer

```python
customer_country = (
    df[['CustomerID', 'Country']]
    .dropna(subset=['CustomerID'])
    .drop_duplicates(subset=['CustomerID'])
)
print(customer_country)
customer_country.to_csv('For Database/customer.csv', index=False)
```

# Product

```python
product = (
    df[['StockCode', 'Description', 'UnitPrice']]
    .dropna(subset=['StockCode'])
    .drop_duplicates(subset=['StockCode'])
)
print(product)
product.to_csv('For Database/product.csv', index=False)
```

# Order Detail

```python
detail_order = (
    df[['InvoiceNo', 'StockCode', 'Quantity', 'UnitPrice']]
    .dropna(subset=['InvoiceNo'])
    .copy()
)
print(detail_order)
detail_order.to_csv('For Database/detail_order.csv', index=False)
```

# Order

```python
order = (
    df[['InvoiceNo', 'InvoiceDate', 'CustomerID']]
    .dropna(subset=['InvoiceNo'])
    .drop_duplicates(subset=['InvoiceNo'])
)
print(order)
```

```python
# Hitung total per invoice
invoice_total = (
    detail_order
    .groupby('InvoiceNo')
    .apply(lambda x: (x['Quantity'] * x['UnitPrice']).sum())
)

# Tambahkan kolom Total ke order
order['Total'] = order['InvoiceNo'].map(invoice_total).fillna(0)

# Simpan ke CSV
order.to_csv('For Database/orders.csv', index=False)
```
