# Import Library

```python
import pandas as pd
```

# Import Data

```python
df = pd.read_csv('order.csv', encoding='utf-8-sig', delimiter=',')
df
```

```python
print(df.dtypes)
```

```python
df = df.astype({'CustomerID' : object})
```

```python
print(df.dtypes)
```

```python
print(df['CustomerID'].nunique())
```

```python
print(df['StockCode'].nunique())
```

```python
df = df.rename(columns={'Description' : 'ProductName'})
```

```python
customer_country = (
    df[['CustomerID', 'Country']]
    .dropna(subset=['CustomerID'])
    .drop_duplicates(subset=['CustomerID'])
)
print(customer_country)
customer_country.to_csv('customer.csv', index=False)
```

```python
product = (
    df[['StockCode', 'ProductName']]
    .dropna(subset=['StockCode'])
    .drop_duplicates(subset=['StockCode'])
)
print(product)
product.to_csv('product.csv', index=False)
```
